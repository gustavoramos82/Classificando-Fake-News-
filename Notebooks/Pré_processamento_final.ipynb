# Pré-processamento final

Além de um pré-processamento feito anteriormente para a pasta prata que foi a remoção de stopwords e lematização, agora vai ser feita os seguintes tratamentos:

- selecionar as noticias dos anos de 2016 e 2017

pois eesse anos tem termos da época e é a maior parte das noticias

- inserindo a label nos dataframes
- juntar os dataframes

Pois veio em dois dataframes diferentes e tem que juntar para fazer a modelagem

- selecionar o campo titulo e label

Pois são apenas esses campos que vão ser utilizados na modelagem

```python
import numpy as np
import pandas as pd
```

```python
true =  pd.read_csv(r'/content/drive/MyDrive/Datasets/Fake news deep/Prata/True_tratado.csv',lineterminator='\n')
true.drop('Unnamed: 0',axis=1,inplace=True)
true.head()
```

```python
fake = pd.read_csv(r'/content/drive/MyDrive/Datasets/Fake news deep/Prata/Fake_tratado.csv',lineterminator='\n')
fake.drop('Unnamed: 0',axis=1,inplace=True)
fake.head()
```

```python
true['date'] = pd.to_datetime(true['date'])
true['ano'] = true['date'].dt.year
true.dropna(inplace=True)
true['ano'] = true['ano'].astype(int)
true.head()
```

```python
fake['date'] = pd.to_datetime(fake['date'])
fake['ano'] = fake['date'].dt.year
fake.dropna(inplace=True)
fake['ano'] = fake['ano'].astype(int)
fake.head()
```

Selecionando o ano de 2016 e 2017 do fake

```python
fake = fake.query('ano>2015 & ano<2018')
```

```python
fake.head()
```

```python
true.shape
```

```python
fake.shape
```

```python
label_true = []
for i in range(true.shape[0]):
  label_true.append(0)
```

```python
len(label_true)
```

```python
true['label'] = label_true
true.head()
```

```python
label_false = []
for i in range(fake.shape[0]):
  label_false.append(1)

len(label_false)
```

```python
fake['label'] = label_false
fake.head()
```

Selecionando as colunas title e label

```python
true = true[['title','label']]
true.head()
```

```python
fake = fake[['title','label']]
fake.head()
```

```python
news = pd.concat([true,fake])
news.head()
```

```python
news.tail()
```

```python
news.reset_index(inplace=True)
news.tail()
```

```python
news.drop('index',axis=1,inplace=True)
news.tail()
```

```python
news.shape
```

```python
news.to_csv('News.csv')
```
