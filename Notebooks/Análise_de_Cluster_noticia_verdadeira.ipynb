# Análise de cluster da noticias verdadeiras

Neste notebook será feita uma análise de cluster das noticias verdadeiras, com a análise será comparada com as outras análises que foi feitas até então.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.decomposition import TruncatedSVD

from yellowbrick.cluster import KElbowVisualizer

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
true = pd.read_csv(r'/content/drive/MyDrive/Datasets/Fake news deep/Prata/True_tratado.csv',lineterminator='\n')
true.drop('Unnamed: 0',axis=1,inplace=True)
true.head()
```

```python
true['date'] = pd.to_datetime(true['date'])
true['ano'] = true['date'].dt.year
true.dropna(inplace=True)
true['ano'] = true['ano'].astype(int)
true.head()
```

```python
tdif = TfidfVectorizer()
matr_title = tdif.fit_transform(true.title)
```

```python
trun_svd = TruncatedSVD(n_components=5)
matr_title = trun_svd.fit_transform(matr_title)
```

```python
kmeans = KMeans()
elbow = KElbowVisualizer(kmeans,k=(2,12))
elbow.fit(matr_title)
elbow.show()
```

```python
elbow_cal = KElbowVisualizer(kmeans,k=(2,12), metric='calinski_harabasz')
elbow_cal.fit(matr_title)
elbow_cal.show()
```

```python
kmeans = KMeans(n_clusters=5)
matr_title = tdif.fit_transform(true.title)
kmeans.fit(matr_title )
true['cluster_tit'] = kmeans.labels_
true.head()
```

```python
text_cluster_0 = " ".join(i for i in true[true['cluster_tit']==0]['title'])
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords,
                      background_color="black").generate(text_cluster_0)
plt.figure( figsize=(12,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
